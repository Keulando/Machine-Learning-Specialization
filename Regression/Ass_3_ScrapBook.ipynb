```python
sales= pd.read_csv('kc_house_data.csv', dtype=dtype_dict)
sales= sales.sort_values(by=['sqft_living','price'])
```

```python
def polynom_df(feature, degrees):
    # degree >=1
    poly_df=pd.DataFrame()
    poly_df['price']=sales['price']
    poly_df['power_1']=feature
    
    #cehck if degree >1
    if degree in degrees > 1:
        for power in range(2, degree+1): #loop over reaming deg
            name = 'power_'+str(power)
            poly_df[name]=poly_df['power_1'].apply(lambda x: x**power)
            
    return poly_df
```

```python
def model_prediction(target):
    for column in poly_df.columns:
        X=poly_df[column].reshape(-1,1)
        y=poly_df[target]
        model=LinearRegression().fit(X,y)
        prediction=model.predict(X)
    return prediction
```

```python
poly_df=polynom_df(sales['sqft_living'],degrees=[1,2,3])
predictions=model_prediction('price')
poly_df
```

```python
def plot_degrees(degrees, feature, target):
    for degree in degrees:
        poly_df=polynom_df(feature,degree=degree)
        predictions=model_prediction(target)
        
    return plt.plot(poly_df.columns[1], poly_df.columns[0], '.',
                        poly_df.columns[1], predictions.reshape(1,-1), '-')
            
```

```python
def Coeff(x,power, debug=False):
    '''Prints intercept and coeff in a better format
    
    args: x= name of model
          power= degree of polynomial function'''
    
    names=["intercept",'power_'+str(power)]
    values=[x.intercept_,x.coef_]
    df=pd.DataFrame(columns=['Name','Value'])
    df['Name']=names
    df['Value']= values
    if debug ==True:
        print('coef is/are: {:.4f}'.format(x.coef_))
        print('intercept is: {:.4f}'.format(x.intercept_))
    return df
```

```python
def poly_df_for_subs(feature, subset, degree):
    # degree >=1
    
    poly_df=pd.DataFrame()
    poly_df['price']=subset['price']
    poly_df['power_1']=subset[str(feature)]
    
    #cehck if degree >1
    if degree > 1:
        for power in range(2, degree+1): #loop over reaming deg
            name = 'power_'+str(power)
            poly_df[name]=poly_df['power_1'].apply(lambda x: x**power)
            
    return poly_df
```

```python
def Coeff2(x, power):
    '''Prints intercept and coeff in a better format
    
    args: x = name of model
      power = degree of polynomial function'''
    
    coef_df=pd.DataFrame()
    coef_df=coef_df.append({"Name":'Intercept', 'Value':x.intercept_}, ignore_index=True)
    coef_df=coef_df.append({'Name':'Power_1','Value':x.coef_[0]}, ignore_index=True)
    
    if power > 1:
        for power in range(2, power+1):
            name= 'Power_' + str(power)
            coef_df=coef_df.append({"Name":name, 'Value':'{:.3e}'.format(x.coef_[power-1])}, ignore_index=True)
    
    return coef_df
```

```python
def prep_model_and_predict(feature, target, degree, plot, *datasets):
    

    
    # part 1: make a df with relevant format and features 
        # degree >=1
    for dataset in datasets:
        poly_df=pd.DataFrame()
        poly_df[str(target)] = dataset[str(target)]
        poly_df['power_1']   = dataset[str(feature)]
    
        #cehck if degree >1
        if degree > 1:
            for power in range(2, degree+1): #loop over reaming deg
                name = 'power_'+str(power)
                poly_df[name]=poly_df['power_1'].apply(lambda x: x**power)
            
    #part 2: make model and predictions
    
        features=list(poly_df.columns[1:])
        X=poly_df[features]
        y=poly_df[str(target)]
        model=LinearRegression().fit(X,y)
        predictions=model.predict(X)
    
    #part 3: put weghts in a nice df
    
        coef_df=pd.DataFrame()
        coef_df=coef_df.append({"Name":'Intercept', 'Value':model.intercept_}, ignore_index=True)
        coef_df=coef_df.append({'Name':'Power_1',   'Value':model.coef_[0]},   ignore_index=True)
    
        if degree > 1:
            for degree in range(2, degree+1):
                name = 'Power_' + str(degree)
                coef_df = coef_df.append({"Name":name, 
                                          'Value':'{:.3e}'.format(model.coef_[degree-1])}, ignore_index=True)
                #print(coef_df)
    #prt 4: plot it
        if plot == True:
            plt.figure()
            plt.plot(poly_df['power_1'], poly_df[str(target)], '.',
                    poly_df['power_1'], predictions, '-')

            plt.xlabel('Square footage, living area')
            plt.ylabel('Price per Sqft')
            plt.ticklabel_format(axis='y', style='sci', scilimits=(-2,2))
            plt.show()
    return coef_df, predictions, model
```
